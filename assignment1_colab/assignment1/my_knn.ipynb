```python
# 展示数据集
import numpy as np
import random
from cs231n.data_utils import load_CIFAR10
import matplotlib.pyplot as plt 

%matplotlib inline 
plt.rcParams['figure.figsize'] = (10.0, 8.0) # 用于这是图像的大小长宽为10,8
plt.rcParams['image.interpolation'] = 'nearest' # 设置图像为最近邻插值
plt.rcParams['image.cmap'] = 'gray' #　设置图像的颜色映射为灰度，用于在图像中显示灰度图
# 用于加载ipython的autoreload扩展模块
# 这是扩展模块在修改后自动重新加载模块
%load_ext autoreload 
%autoreload 2 
```

```python
# 加载数据集
cifar10_dir = 'cs231n/datasets/cifar-10-batches-py'
# 清除变量
try:
    del X_train, y_train
    del X_test, y_test
    print("删除成功\n")
except:
    pass

X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

print('training data shape:',X_train.shape)
print('training lable shape:',y_train.shape)
print('test data shape:',X_test.shape)
print('test lable shape:',y_test.shape)

```

```python
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
sample_per_class = 7
for y, cls in enumerate(classes):
    """ 0 plane
        1 car
        2 bird
        3 cat
        4 deer
        5 dog
        6 frog
        7 horse
        8 ship
        9 truck"""
    # 取出测试集中y类的标签的索引
    idxs = np.flatnonzero(y_train == y)
    # 随机取出来sample_per_class个
    idxs = np.random.choice(idxs, sample_per_class, replace = False)
    # 开始画图
    
    for i, idx in enumerate(idxs):
        #　设置图形的位置
        plt_idx = i * num_classes + y + 1
        plt.subplot(sample_per_class, num_classes, plt_idx)
        
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)
plt.show()
        
        
    
```
